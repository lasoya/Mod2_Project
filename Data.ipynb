## Overall USDA Mangos Import Data - Unit Price

## Working with Mango Data In Smaller Groups - Unit Price

```python
# split up the overall data into smaller dataframes each representing each type of mango product and their unit prices over time
unit_price_dried, unit_price_fresh, unit_price_frozen, unit_price_juice, unit_price_prepared = pd.DataFrame(unit_price[:10]).reset_index().drop(['index'], axis=1), pd.DataFrame(unit_price[11:21]).reset_index().drop(['index'], axis=1), pd.DataFrame(unit_price[22:32]).reset_index().drop(['index'], axis=1), pd.DataFrame(unit_price[33:43]).reset_index().drop(['index'], axis=1), pd.DataFrame(unit_price[44:]).reset_index().drop(['index'], axis=1)
```

```python
# sort values of the dried mango unit price data
unit_price_dried = unit_price_dried.sort_values('Market year*', ascending=True)
```

```python
unit_price_dried
```

```python
# plot the data to see the trend per year
unit_price_dried.plot(x='Market year*', y='Avg')
```

```python
# sort values of the fresh mango unit price data
unit_price_fresh = unit_price_fresh.sort_values('Market year*', ascending=True)
```

```python
unit_price_fresh
```

```python
# plot the data to see the trend per year
unit_price_fresh.plot(x='Market year*', y='Avg')
```

```python
# write to csv file
unit_price_fresh.to_csv('unit_price_fresh.csv')
unit_price_dried.to_csv('unit_price_dried.csv')
```
